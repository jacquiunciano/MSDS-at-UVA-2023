```python
OHCO = ['book_title', 'chap_num', 'para_num', 'sent_num', 'token_num']
BOOKS = OHCO[:1]
CHAPS = OHCO[:2]
PARAS = OHCO[:3]
SENTS = OHCO[:4]
```

```python
salex_csv = f'{data_home}\\salex_nrc.csv'
emo_cols = "anger anticipation disgust fear joy sadness surprise trust sentiment".split()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly_express as px
from IPython.display import display, HTML

sns.set()
```

```python
CORPUS = pd.read_csv(f"{output_dir}\\CORPUS.csv").set_index(OHCO)
LIB = pd.read_csv(f"{output_dir}\\LIB_LABELS.csv").set_index(BOOKS)[["series","year","author"]]
TFIDF = pd.read_csv(f"{output_dir}\\TFIDF_chaps.csv").set_index(CHAPS)
VOCAB = pd.read_csv(f"{output_dir}\\VOCAB.csv").set_index('term_str')
```

```python
df = TFIDF.groupby(CHAPS)\
    .mean()\
    .stack()\
    .to_frame('mean_tfidf')
df.index.names = CHAPS+['term_str']
BOW = df.query("mean_tfidf != 0")
print(BOW.shape)
BOW.head()
```

```python
SALEX = pd.read_csv(salex_csv).set_index('term_str')
SALEX.columns = [col.replace('nrc_','') for col in SALEX.columns]
```

```python
SALEX.head()
```

```python
V = pd.concat([VOCAB.reset_index().set_index('term_str'), SALEX], join='inner', axis=1)
V.sample(10)
```

```python
B = BOW.join(V[['max_pos'] + emo_cols], on='term_str', rsuffix='_v').dropna()
B.sample(10)
```

### Weighing by TFIDF

```python
for col in emo_cols:
    B[col] = B[col] * B.mean_tfidf

B = B.reset_index().merge(LIB.reset_index(), on="book_title", how="left").set_index(CHAPS)

B.head()
```

```python
import matplotlib.ticker as ticker

def plot_sentiments(df, emo='sentiment', index_dtype='str'):
    if index_dtype == 'str':
        df.index = df.index.astype(str)
    elif index_dtype == 'int':
        df.index = df.index.astype(int)
        
    FIG = dict(figsize=(25, 5), legend=True, fontsize=14, rot=45)
    ax = df[emo].plot(**FIG)
    
    if index_dtype == 'int':
        ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
    
    plt.show()
```

```python
plot_sentiments(EMO_YEARS, emo_cols[:-1], index_dtype='int')
```

```python
vis_setup = B.groupby(CHAPS)[emo_cols].mean().reset_index()\
    .merge(LIB.reset_index(), on="book_title", how="left").set_index(CHAPS)
vis_setup.head()
```

```python
def plot_emo_series(df, emotion='fear'):
    plt.figure(figsize=(10, 6))  
    
    for label, grp in df.groupby('series'):
        new_df = grp.groupby(["book_title"])[emotion].mean().to_frame(emotion)
        new_df["book_num"] = new_df.index.str.split("_").str[0]
        plt.plot(new_df.book_num, new_df[emotion], label=label)

    plt.title(f'{emotion.capitalize()} by Series')
    plt.xlabel('Book')
    plt.ylabel(emotion.capitalize())
    plt.legend(title='Series')
    plt.grid(True)

    plt.show()
```

```python
plot_emo_series(vis_setup)
```

```python
plot_emo_series(vis_setup, emotion="trust")
```

```python
plot_emo_series(vis_setup, emotion="sadness")
```
