```python
## get win/loss records in 2021-2022 season
## inner, full, right, left join
## inner = deletes everything that does not match
## full = keep everything that does not match and fill with NA
## right = keep all of rows in JOIN and delete everything that doesn't match in FROM
## left = keep all of rows in FROM and delete everything that doesn't match in JOIN
## tg and g are aliases so we can ref cols
## ex: tg.colname vs g.colname
mq = '''
SELECT tg.game_id, tg.Team_Abbrev, tg.Opponent_Abbrev, tg.Team_Score,
    sg.Team_Score AS Opponent_Score,
    (tg.Team_Score > sg.Team_Score) AS win,
    (tg.Team_Score < sg.Team_Score) AS loss
FROM team_game tg
INNER JOIN games g
    ON tg.game_id=g.game_id
INNER JOIN team_game sg
    ON tg.game_id=sg.game_id AND tg.Opponent_Abbrev=sg.Team_Abbrev
WHERE g.season=2022
'''
```

```python
pd.read_sql_query(mq, nba)
```
