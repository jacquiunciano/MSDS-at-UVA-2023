## Problem 14
Write `pandas` code that replicates the output of the following SQL code:
```
SELECT size, type, premiums AS insurance, percent_female FROM whpps
WHERE industry = 'Hospitals' AND premium_change='Smaller'
ORDER BY percent_female DESC;
```
For each of these queries, your feature names might be different from the ones listed in the query, depending on the names you chose in problem 3.
[2 points]

````python
## SQL version
wha_cursor.execute("SELECT Employee_Size, Organization_Type_Profit, Health_Insur_Type AS insurance, Sex_F \
FROM wha \
WHERE Industry = 'Hospital Worksites' AND Personal_Health_Insur='Smaller' \
ORDER BY Sex_F DESC")
cur_df = wha_cursor.fetchall()
pd.DataFrame(cur_df)
````

````python
## pandas version
df.loc[:,["Employee_Size", "Organization_Type_Profit", "Health_Insur_Type", "Sex_F"]]\
.loc[np.where((df["Industry"] == 'Hospital Worksites') & (df["Personal_Health_Insur"]=='Smaller'))]\
.sort_values("Sex_F", ascending=False)
````

## Problem 15
Write `pandas` code that replicates the output of the following SQL code:
```
SELECT industry, 
    AVG(percent_female) as percent_female, 
    AVG(percent_under30) as percent_under30,
    AVG(percent_over60) as percent_over60
FROM whpps
GROUP BY industry
ORDER BY percent_female DESC;
```
[2 points]

````python
## SQL version
wha_cursor.execute("SELECT Industry, AVG(Sex_F) AS percent_female, \
AVG(Under_30) AS percent_under30, AVG(Plus_60) AS percent_over60 \
FROM wha \
GROUP BY Industry \
ORDER BY percent_female DESC")
cur_df = wha_cursor.fetchall()
pd.DataFrame(cur_df)
````

````python
## pandas version
df.loc[:,["Industry", "Sex_F", "Under_30", "Plus_60"]]\
.groupby(by="Industry").mean()\
.sort_values("Sex_F", ascending=False)
````

## Problem 16
Write `pandas` code that replicates the output of the following SQL code:
```
SELECT gender_balance, premiums, COUNT(*)
FROM whpps
GROUP BY gender_balance, premiums
HAVING gender_balance is NOT NULL and premiums is NOT NULL;
```
[2 points]

````python
## SQL version
wha_cursor.execute("SELECT gender_balance, Health_Insur_Type, COUNT(*) \
FROM wha \
GROUP BY gender_balance, Health_Insur_Type \
HAVING gender_balance is NOT NULL and Health_Insur_Type is NOT NULL")
cur_df = wha_cursor.fetchall()
pd.DataFrame(cur_df)
````
