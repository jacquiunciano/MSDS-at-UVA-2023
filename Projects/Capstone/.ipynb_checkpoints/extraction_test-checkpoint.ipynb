```python
import fitz
```

```python
doc = fitz.open('usvbell.pdf')
```

```python
text = "" 
for page in doc: 
   text+=page.get_text() 
```

```python
start = text.find("Summary of the Evidence")
end = text.find("a. Counts 1 through 3")
```

```python
extracted = text[start:end]
```

```python
extracted = extracted.replace('\n', '')
```

```python
extracted
```
